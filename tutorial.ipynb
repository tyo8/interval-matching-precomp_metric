# Tutorial - Fast cohomological cycle matching

This is a tutorial to illustrate how to implement the code on this repository. We will do this by matching two circles with the same radies and centers shifted by some distance that one can control.

**Remark**: you have to include the compiled versions of ripser and ripser-image in this folder in order for this notebook to work properly. 

```python
# to import the utils correctly
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
# import the functions
from match.utils_PH import *
from match.utils_plot import *
from match.utils_data import *
```

```python
# generate our sample: two circles with equal radius and centres shifted
X = sample_circle(x0 = 0, r = 1, N = 50, dim = 2, noise_scale = .1) 
Y = sample_circle(x0 = 0.5, r = 1, N = 50, dim = 2, noise_scale = .1) 

plot_two_point_clouds(X,Y)
```

```python
matched_X_Y, affinity_X_Y, bars_reps_X, bars_reps_Y  = matching(X, Y, dim = 1, affinity_method = 'A', check_Morse = True)    

bars_X, reps_X, tight_reps_X = bars_reps_X
bars_Y, reps_Y, tight_reps_Y = bars_reps_Y

show_matches(X, Y, matched_X_Y, affinity_X_Y, tight_reps_X, tight_reps_Y, zoom_factor = 3)
```
